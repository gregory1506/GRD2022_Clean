```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
locA_a1_rain = pd.read_excel("Country A.xlsx", sheet_name=2, skiprows=3)
locA_a1_rain = locA_a1_rain.melt(id_vars=['DATE','Mth'], var_name='year', value_name='rain')
locA_a1_rain = locA_a1_rain[locA_a1_rain.DATE <= 365]
locA_a1_rain['DT'] = pd.to_datetime(locA_a1_rain['year'] * 1000 + locA_a1_rain['DATE'], format='%Y%j')
locA_a1_rain = locA_a1_rain.rename(columns={'DATE':'doy', 'Mth':'Month', 'DT':'Date'})
locA_a1_rain
```

```python
locA_a1_avgtemp = pd.read_excel("Country A.xlsx", sheet_name=3, skiprows=3)
locA_a1_avgtemp = locA_a1_avgtemp.melt(id_vars=['DATE','Mth'], var_name='year', value_name='Avg_Temp')
locA_a1_avgtemp = locA_a1_avgtemp[locA_a1_avgtemp.DATE <= 365]
locA_a1_avgtemp['DT'] = pd.to_datetime(locA_a1_avgtemp['year'] * 1000 + locA_a1_avgtemp['DATE'], format='%Y%j')
locA_a1_avgtemp = locA_a1_avgtemp.rename(columns={'DATE':'doy', 'Mth':'Month', 'DT':'Date'})
locA_a1_avgtemp
```

```python
locA_a1_mintemp = pd.read_excel("Country A.xlsx", sheet_name=4, skiprows=2)
locA_a1_mintemp = locA_a1_mintemp.melt(id_vars=['DATE','Mth'], var_name='year', value_name='Min_Temp')
locA_a1_mintemp = locA_a1_mintemp[locA_a1_mintemp.DATE <= 365]
locA_a1_mintemp['DT'] = pd.to_datetime(locA_a1_mintemp['year'] * 1000 + locA_a1_mintemp['DATE'], format='%Y%j')
locA_a1_mintemp = locA_a1_mintemp.rename(columns={'DATE':'doy', 'Mth':'Month', 'DT':'Date'})
locA_a1_mintemp
```

```python
locA_a1_maxtemp = pd.read_excel("Country A.xlsx", sheet_name=5, skiprows=2)
locA_a1_maxtemp = locA_a1_maxtemp.melt(id_vars=['DATE','Mth'], var_name='year', value_name='Max_Temp')
locA_a1_maxtemp = locA_a1_maxtemp[locA_a1_maxtemp.DATE <= 365]
locA_a1_maxtemp['DT'] = pd.to_datetime(locA_a1_maxtemp['year'] * 1000 + locA_a1_maxtemp['DATE'], format='%Y%j')
locA_a1_maxtemp = locA_a1_maxtemp.rename(columns={'DATE':'doy', 'Mth':'Month', 'DT':'Date'})
locA_a1_maxtemp
```

```python
df = locA_a1_rain.merge(locA_a1_avgtemp[['Avg_Temp']], left_index=True, right_index=True, how='left')
df = df.merge(locA_a1_mintemp[['Min_Temp']], left_index=True, right_index=True, how='left')
df = df.merge(locA_a1_maxtemp[['Max_Temp']], left_index=True, right_index=True, how='left')
df['area'] = 1
df['country'] = "A"
df
```

```python
locA_a2_rain = pd.read_excel("Country A.xlsx", sheet_name=7, skiprows=4)
locA_a2_rain = locA_a2_rain.melt(id_vars=['DAY','MONTH'], var_name='year', value_name='rain')
locA_a2_rain = locA_a2_rain[(locA_a2_rain.DAY <= 365) & (locA_a2_rain.year >= 1990)]
locA_a2_rain['DT'] = pd.to_datetime(locA_a2_rain['year'] * 1000 + locA_a2_rain['DAY'], format='%Y%j')
locA_a2_rain = locA_a2_rain.rename(columns={'DAY':'doy', 'MONTH':'Month', 'DT':'Date'}).reset_index(drop=True)
locA_a2_rain
```

```python
locA_a2_avgtemp = pd.read_excel("Country A.xlsx", sheet_name=8, skiprows=4)
locA_a2_avgtemp = locA_a2_avgtemp.melt(id_vars=['DAY','MONTH'], var_name='year', value_name='Avg_Temp')
locA_a2_avgtemp = locA_a2_avgtemp[locA_a2_avgtemp.DAY <= 365]
locA_a2_avgtemp['DT'] = pd.to_datetime(locA_a2_avgtemp['year'] * 1000 + locA_a2_avgtemp['DAY'], format='%Y%j')
locA_a2_avgtemp = locA_a2_avgtemp.rename(columns={'DAY':'doy', 'MONTH':'Month', 'DT':'Date'})
locA_a2_avgtemp
```

```python
locA_a2_mintemp = pd.read_excel("Country A.xlsx", sheet_name=10, skiprows=4)
locA_a2_mintemp = locA_a2_mintemp.melt(id_vars=['DATE','MONTH'], var_name='year', value_name='Min_Temp')
locA_a2_mintemp = locA_a2_mintemp[locA_a2_mintemp.DATE <= 365]
locA_a2_mintemp['DT'] = pd.to_datetime(locA_a2_mintemp['year'] * 1000 + locA_a2_mintemp['DATE'], format='%Y%j')
locA_a2_mintemp = locA_a2_mintemp.rename(columns={'DATE':'doy', 'MONTH':'Month', 'DT':'Date'})
locA_a2_mintemp
```

```python
locA_a2_maxtemp = pd.read_excel("Country A.xlsx", sheet_name=9, skiprows=4)
locA_a2_maxtemp = locA_a2_maxtemp.melt(id_vars=['DATE','MONTH'], var_name='year', value_name='Max_Temp')
locA_a2_maxtemp = locA_a2_maxtemp[locA_a2_maxtemp.DATE <= 365]
locA_a2_maxtemp['DT'] = pd.to_datetime(locA_a2_maxtemp['year'] * 1000 + locA_a2_maxtemp['DATE'], format='%Y%j')
locA_a2_maxtemp = locA_a2_maxtemp.rename(columns={'DATE':'doy', 'MONTH':'Month', 'DT':'Date'})
locA_a2_maxtemp
```

```python
df2 = locA_a2_rain.merge(locA_a2_avgtemp[['Avg_Temp']], left_index=True, right_index=True, how='left')
df2 = df2.merge(locA_a2_mintemp[['Min_Temp']], left_index=True, right_index=True, how='left')
df2 = df2.merge(locA_a2_maxtemp[['Max_Temp']], left_index=True, right_index=True, how='left')
df2['area'] = 2
df2['country'] = "A"
df2['year'] = df2.year.astype(int)
df2['rain'] = df2.rain.apply(lambda x: np.nan if x == "N.R" else float(x))
df2['Avg_Temp'] = df2.Avg_Temp.apply(lambda x: str(x).replace("..","."))
df2['Avg_Temp'] = df2.Avg_Temp.apply(lambda x: np.nan if x == "n.a" else float(x))
df2['Min_Temp'] = df2.Min_Temp.apply(lambda x: np.nan if (x == "n.a " or x=="n.a") else float(x))
df2['Max_Temp'] = df2.Max_Temp.apply(lambda x: str(x).replace("..","."))
df2['Max_Temp'] = df2.Max_Temp.apply(lambda x: np.nan if (x == "n.a " or x=="n.a") else float(x))
df2
```

```python
df2.to_parquet("CountryA_Area2_cleaned_merged.parquet")
# df1.to_parquet("CountryA_Area1_cleaned_merged.parquet")
```
