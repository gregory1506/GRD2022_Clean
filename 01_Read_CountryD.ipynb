```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
locD_a1_temps = pd.read_excel("Country D.xlsx", sheet_name=3, skiprows=3).dropna().reset_index(drop=True).rename(columns={'Year':'year'})
locD_a1_temps = locD_a1_temps.melt(id_vars='year',var_name='Month',value_name='Avg_Temp')
months = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}
locD_a1_temps.Month = locD_a1_temps.Month.map(months)
locD_a1_temps['doy'] = locD_a1_temps.apply(lambda row: datetime(int(row.year), int(row.Month), 1).timetuple().tm_yday, axis=1)
locD_a1_temps
```

```python
months2 = {
    "Jan": 1,
    "Feb": 2,
    "Mar": 3,
    "Apr": 4,
    "May": 5,
    "Jun": 6,
    "Jul": 7,
    "Aug": 8,
    "Sept": 9,
    "Oct ": 10,
    "Nov ": 11,
    "Dec": 12
}
locD_a1_rain = pd.read_excel("Country D.xlsx", sheet_name=2, skiprows=4).iloc[:12].melt(id_vars='Month', var_name='year', value_name='rain').reset_index(drop=True)
locD_a1_rain.Month = locD_a1_rain.Month.map(months2)
locD_a1_rain['doy'] = locD_a1_rain.apply(lambda row: datetime(row.year, row.Month, 1).timetuple().tm_yday, axis=1)
locD_a1_rain.year = locD_a1_rain.year.astype(int)
locD_a1_rain.rain = locD_a1_rain.rain.apply(lambda x: float(x) if x not in ["n.a"] else np.nan)
locD_a1_rain.sort_values(by=['year','Month','doy'])
```

```python
df = pd.merge(locD_a1_temps, locD_a1_rain, left_on=['year', 'Month','doy'], right_on=['year', 'Month','doy'], how='right')
df['area'] = 1
df['country'] = "D"
df['Date'] = df.apply(lambda row: datetime(row.year, row.Month, 1).strftime("%Y-%m-%d"), axis=1)
df = df[['doy','Month','year','rain','Date','Avg_Temp','area','country']]
df
```
